#### Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

#### Load the dataset

```python
waterdf = pd.read_csv("water_potability.csv")
waterdf.head()
```

#### Remove all the rows that contain null values

```python
waterdf = waterdf.dropna()
waterdf.isnull().sum()
```

### 0 (unfit) and 1 (potable) for consumption

#### 1. Distribution of 0 and 1 in the Potability column

```python
plt.figure(figsize=(10, 5))
sns.countplot(data = waterdf, x = 'Potability')
plt.title("Distribution of Unsafe and Safe Water")
```

So this dataset is not balanced because samples of 0s are more than 1s

 #### 2. Create a histogram chart to visualize the distribution of ph

```python
px.histogram(waterdf, x = "ph", color = "Potability", title= "Factors Affecting Water Quality: PH")
```

The ph column represents the ph value of the water which is an important factor in evaluating the acid-base balance of the water. The pH value of drinking water should be between 6.5 and 8.5

 #### 3. Create a histogram chart to visualize the distribution of water hardness

```python
px.histogram(waterdf, x = "Hardness", color = "Potability", title= "Factors Affecting Water Quality: Hardness")
```

The hardness of water usually depends on its source, but water with a hardness of 120-200 milligrams is drinkable

 #### 4. Create a histogram chart to visualize the distribution of total dissolved solids in water

```python
px.histogram(waterdf, x = "Solids", color = "Potability", title= "Factors Affecting Water Quality: Solids")
```

All organic and inorganic minerals present in water are called dissolved solids. Water with a very high number of dissolved solids is highly mineralized

 #### 5. Create a histogram chart to visualize the distribution of chloramine in water

```python
px.histogram(waterdf, x = "Chloramines", color = "Potability", title= "Factors Affecting Water Quality: Chloramines")
```

Chloramine and chlorine are disinfectants used in public water systems

 #### 6. Create a histogram chart to visualize the distribution of sulfate in water

```python
px.histogram(waterdf, x = "Sulfate", color = "Potability", title= "Factors Affecting Water Quality: Sulfate")
```

They are substances naturally present in minerals, soil and rocks. Water containing less than 500 milligrams of sulfate is safe to drink

 #### 7. Create a histogram chart to visualize the distribution of water conductivity

```python
px.histogram(waterdf, x = "Conductivity", color = "Potability", title= "Factors Affecting Water Quality: Conductivity")
```

Water is a good conductor of electricity, but the purest form of water is not a good conductor of electricity. Water with an electrical conductivity of less than 500 is drinkable

 #### 8. Create a histogram chart to visualize the distribution of organic carbon in water

```python
px.histogram(waterdf, x = "Organic_carbon", color = "Potability", title= "Factors Affecting Water Quality: Organic Carbon")
```

Organic carbon comes from the breakdown of natural organic materials and synthetic sources. Water containing less than 25 milligrams of organic carbon is considered safe to drink

 #### 9. Create a histogram chart to visualize the distribution of trihalomethanes or THMs in water

```python
px.histogram(waterdf, x = "Trihalomethanes", color = "Potability", title= "Factors Affecting Water Quality: Trihalomethanes")
```

THMs are chemicals found in chlorine-treated water. Water containing less than 80 milligrams of THMs is considered safe to drink

 #### 10. Create a histogram chart to visualize the distribution of turbidity in water

```python
px.histogram(waterdf, x = "Turbidity", color = "Potability", title= "Factors Affecting Water Quality: Turbidity")
```

The turbidity of water depends on the number of solids present in suspension. Water with a turbidity of fewer than 5 milligrams is considered drinkable
