```python
# Univariate Analysis
```

```python
df.columns
```

```python
sns.histplot(df['total_bill'], kde=True)
plt.title('Distribution of Total Bill')
plt.show()
```

```python
sns.boxplot(x=df['total_bill'])
plt.title('Boxplot of Total Bill')
plt.show()
```

```python
sns.histplot(df['tip'], kde = True)
plt.title('Distribution of Tip')
plt.show()
```

```python
sns.boxplot(x=df['tip'])
plt.title('Boxplot of Tip')
plt.show()
```

```python
sns.countplot(x='size', data=df)
plt.title('Distribution of Table Size')
plt.show()
```

```python
sns.countplot(x='sex', data=df)
plt.title('Distribution of Gender')
plt.show()
```

```python
sns.countplot(x='smoker', data=df)
plt.title('Distribution of Smoker')
plt.show()
```

```python
sns.countplot(x='day', data=df)
plt.title('Distribution of Days')
plt.show()
```

```python
sns.countplot(x='time', data=df)
plt.title('Lunch vs Dinner')
plt.show()
```

```python
# Bivariate Analysis
```

```python
sns.scatterplot(x='total_bill', y='tip',data=df)
plt.title('Total Bill vs Tip')
plt.show()
```

```python
df[['total_bill','tip']].corr()
```

```python
sns.boxplot(x='sex', y='tip', data=df)
plt.title('Tip by Gender')
plt.show()
```

```python
sns.boxplot(x='smoker', y='tip', data=df)
plt.title('Tip by Smoking Status')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
plt.title('Total Bill by Day')
plt.show()
```

```python
sns.countplot(x='sex', hue='smoker', data=df)
plt.title('Smoker by Gender')
plt.show()
```

```python
sns.countplot(x='smoker', hue='sex', data=df)
plt.title('Gender by Smoker')
plt.show()
```

```python
sns.countplot(x='day', hue='time', data=df)
plt.title('Day vs Meal Time')
plt.show()
```

```python
sns.countplot(x='time', hue='day', data=df)
plt.title('Meal vs Day Time')
plt.show()
```

```python
# Multivariate Analysis
```

```python
corr = df.corr(numeric_only = True)
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.boxplot(x='day',y='tip',hue='time',data=df)
plt.title('Tip by Day and Meal Time')
plt.show()
```

```python
sns.boxplot(x='sex', y='total_bill', hue='smoker', data=df)
plt.title('Total Bill by Gender and Smoking Status')
plt.show()
```

```python
sns.pairplot(
    df,
    hue='sex',
    vars=['total_bill','tip','size']
)
plt.show()
```

Final Insights
- Higher total bill --> Higher tips
- Weekend dinner generate maximum revenue
- Larger group spend more
- Smokers and males slightly tip more
- Lunch customers are fewer and spend less
